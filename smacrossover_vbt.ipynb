```python
import pandas as pd
import vectorbt as vbt
from trader.trader import Trader
```

```python
trader = Trader()
trader.add_broker("mt5")

df = trader.broker._fetch_broker_data("EURUSD", start="2024-01-01", end="2024-06-30", granularity="1min", price='B')
```

```python
df.to_csv("EURUSD_2024_1min.csv", index=True)
```

```python
df2 = pd.read_csv("EURUSD_2024_1min.csv")
df2["datetime"] = pd.to_datetime(df2["datetime"])
df2.set_index("datetime", inplace=True)
df2
```

```python
price = df.close

fast_ma = vbt.MA.run(price, 50, short_name='fast_ma')
slow_ma = vbt.MA.run(price, 200, short_name='slow_ma')

entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)

portfolio = vbt.Portfolio.from_signals(price, entries, exits, init_cash=1000)
```

```python
portfolio.total_profit()
```

```python
portfolio.plot()
```
